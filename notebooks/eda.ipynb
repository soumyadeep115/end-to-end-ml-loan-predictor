```python
# Define bins from 1L to 7L (in rupees)
bins = [100000, 200000, 300000, 400000, 500000, 600000, 700000]
labels = ['1-2L', '2-3L', '3-4L', '4-5L', '5-6L', '6-7L']

# Create a new column with binned income
df['Income_Bin'] = pd.cut(df['ApplicantIncome_Rupees'], bins=bins, labels=labels, include_lowest=True)

# Countplot showing loan status by income bin
plt.figure(figsize=(8,5))
sns.countplot(x='Income_Bin', hue='Loan_Status', data=df)
plt.title("Loan Status by Applicant Income")
plt.xlabel("Applicant Income (in Lakh Rupees)")
plt.ylabel("Count")
plt.show()

# Optional: Approval rates per bin
approval_rates = df.groupby('Income_Bin')['Loan_Status'].value_counts(normalize=True).unstack()
print(approval_rates)
```

```python
# Get all unique loan terms
terms = df['Loan_Amount_Term'].unique()

# Loop through each term and plot pie chart
for t in terms:
    term_data = df[df['Loan_Amount_Term'] == t]['Loan_Status'].value_counts()
    
    plt.figure(figsize=(5,5))
    plt.pie(term_data, labels=term_data.index, autopct='%1.1f%%', colors=['orange','blue'])
    plt.title(f"Loan Status for Loan Amount Term = {t} months")
    plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Credit_History", hue="Loan_Status")
plt.title("Loan Status by Credit History")
plt.show()

# Approval rates
print(df.groupby("Credit_History")["Loan_Status"].value_counts(normalize=True))
```

```python
df['CoApplicantIncome_Rupees'] = df['CoapplicantIncome'] * 1000
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

numerical_features = ['ApplicantIncome_Rupees', 'CoApplicantIncome_Rupees', 'LoanAmount_Rupees']
target = 'Loan_Status'  # Yes/No

for feature in numerical_features:
    plt.figure(figsize=(8,5))
    
    # Violin plot
    sns.violinplot(x=target, y=feature, data=df, inner='quartile', palette='Pastel1')
    plt.title(f'{feature} vs {target} (Violin Plot)')
    plt.show()
    
    # Swarm plot (optional, shows individual points)
    plt.figure(figsize=(8,5))
    sns.swarmplot(x=target, y=feature, data=df, palette='Set2', alpha=0.7)
    plt.title(f'{feature} vs {target} (Swarm Plot)')
    plt.show()
```

```python
# Total Income in Rupees
df['TotalIncome_Rupees'] = df['ApplicantIncome_Rupees'] + df['CoApplicantIncome_Rupees']

# Debt to Income Ratio
df['DebtToIncomeRatio'] = df['LoanAmount_Rupees'] / df['TotalIncome_Rupees']
```
